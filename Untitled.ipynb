```python
sample_rate = 48000

hz = np.linspace(0, sample_rate / 2., 1000)
mel = 1127. * np.log(1. + (hz / 700.))

def to_pi(x):
    return x / x[-1] * np.pi
```

```python
def warped_freq(alpha, nfft):
    """based on replacing the FT kernel, i.e. exp(-1j*omega_k * n) with first-order all-pass system"""
    k = np.arange(nfft//2+1)
    omega_k = 2 * k * np.pi / nfft
    H = (np.exp(-1j*omega_k)-alpha) / (1-alpha*np.exp(-1j*omega_k))
    omega_tilde_k = -np.arctan2(np.imag(H), np.real(H))
    return omega_tilde_k
```

```python
alpha = 0.58

omega = np.linspace(0, np.pi, 1024 // 2 + 1)
omega_tilde = warped_freq(alpha, frame_length)

# omega_tilde = np.arctan2(((1 - alpha**2.) * np.sin(omega))
#                         /
#                         ((1 + alpha**2.) * np.cos(omega) - 2 * omega))

plt.plot(to_pi(hz), to_pi(mel), c='r')
plt.plot(omega, omega_tilde, c='b')
```

```python
def plot_alpha_to_mel(alpha, sample_rate):
    frame_length = 1024
    nfft_half = frame_length // 2 + 1

    hz = np.linspace(0, sample_rate / 2., nfft_half)    
    mel = 1127. * np.log(1. + (hz / 700.))
    mel = mel / mel.max() * np.pi
    
    omega = np.linspace(0, np.pi, nfft_half)
    H = (np.exp(-1j * omega) - alpha) / (1 - alpha * np.exp(-1j * omega))
    warped_omega = -np.arctan2(np.imag(H), np.real(H))
    
    plt.figure(figsize=(8, 8))
    plt.plot(omega, mel, label=f'mel (sr={sample_rate})')
    plt.plot(omega, warped_omega, label=f'alpha={alpha}')
    plt.legend(loc='lower right')
    plt.xticks([0, np.pi/2., np.pi], [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])
    plt.yticks([0, np.pi/2., np.pi], [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$'])
    plt.show()
```

# From SPTK wav

```python
sr, x = wavfile.read(f'fig/{file_id}/{file_id}.wav')
assert sr == 16000
x = x.astype(np.float64)
print(x.shape)
```

```python
# Note that almost all of pysptk functions assume input array is C-contiguous and np.float64 element type
frames = librosa.util.frame(x, frame_length=frame_length, hop_length=hop_length).astype(np.float64).T

# Windowing
frames *= pysptk.blackman(frame_length)

assert frames.shape[1] == frame_length 
```

```python
mc = pysptk.mcep(frames, order, alpha)[65:]
logH = pysptk.mgc2sp(mc, alpha, 0.0, frame_length).real
librosa.display.specshow(logH.T,
                         sr=sr, hop_length=hop_length, x_axis="time", y_axis="linear", cmap='viridis')
colorbar()
plt.savefig(f'fig/{file_id}/sp_mag_sptk.pdf')
mc.shape
```

```python
print(f'{np.min(mc[:, 0]):.4f} {np.max(mc[:, 0]):.4f}')
print(f'{np.min(mc[:, 1]):.4f} {np.max(mc[:, 1]):.4f}')
print(f'{np.min(mc[:, 2]):.4f} {np.max(mc[:, 2]):.4f}')
print(f'{np.min(mc[:, 3]):.4f} {np.max(mc[:, 3]):.4f}')
librosa.display.specshow(np.abs(mc).T, sr=sr, hop_length=hop_length, x_axis="time", y_axis="linear", cmap='viridis')
colorbar()
plt.savefig(f'fig/{file_id}/mc_sptk.pdf')
```
